```python
import os
import math
from pathlib import Path

import mdtraj as md
import numpy as np
from tqdm import tqdm

from afmfold.domain import get_domain_pairs, compute_domain_distance
from afmfold.runner.batch_inference import inference_jsons
from afmfold.utils import add_arg_to_json, suppress_output, k_nearest_weighted_average

def conditional_generation_local(
    target_domain_distance,
    domain_pairs,
    out_dir,
    name,
    json_file,
    input_dict=None,
    base_seed=0,
    t_start=0.54,
    y_max=0.60,
    max_trial=3,
    mse_threshold=10.0,
    suppress=True,
    prev_restraints=None,
    prev_predictions=None,
):
    if input_dict is None:
        input_dict = {}

    if prev_restraints is not None and prev_predictions is not None:
        restraint_domain_distance = k_nearest_weighted_average(
            prev_restraints,
            prev_predictions,
            np.copy(target_domain_distance[None, :]),
            k=6,
        )[0]
    else:
        restraint_domain_distance = target_domain_distance.copy()

    seed = base_seed
    successful_cif_list = []
    failed_cif_list = []

    for _ in range(max_trial):
        kwargs = {
            "sample_diffusion.N_sample": 1,
            "guidance_kwargs.t_start": t_start,
            "guidance_kwargs.manual": restraint_domain_distance,
            "guidance_kwargs.scaling_kwargs.func_type": "sigmoid",
            "guidance_kwargs.scaling_kwargs.y_max": y_max,
            "guidance_kwargs.domain_pairs": domain_pairs,
        }

        while len(list(Path(out_dir, name, f"seed_{seed}").glob('*'))) > 0:
            seed += 1

        seeds = (seed,)
        with suppress_output(suppress=suppress):
            inference_jsons(json_file, out_dir, seeds=seeds, **kwargs)

        output_cifs = list(Path(out_dir, name, f"seed_{seeds[0]}", "predictions").glob('*.cif'))
        if len(output_cifs) != 1:
            err_root = Path(out_dir, 'ERR')
            err_hint = err_root if err_root.exists() else Path(out_dir)
            raise AssertionError(f"No CIFs in {out_dir}/{name}/seed_{seeds[0]}. Check {err_hint}")
        output_cif = str(output_cifs[0])

        output_jsons = list(Path(out_dir, name, f"seed_{seeds[0]}", "predictions").glob('*.json'))
        if len(output_jsons) != 1:
            raise AssertionError(f"No JSONs in {out_dir}/{name}/seed_{seeds[0]}")
        output_json = str(output_jsons[0])

        pred_traj = md.load(output_cif)
        pred_domain_distance = np.zeros((len(domain_pairs),))
        for k, (d1, d2) in enumerate(domain_pairs):
            d = compute_domain_distance(pred_traj, d1, d2)
            pred_domain_distance[k] = d.item()
        pred_domain_distance *= 10.0

        if prev_restraints is not None and prev_predictions is not None:
            prev_restraints = np.concatenate([prev_restraints, restraint_domain_distance[None, :]], axis=0)
            prev_predictions = np.concatenate([prev_predictions, restraint_domain_distance[None, :]], axis=0)

        mse = np.sum((target_domain_distance - pred_domain_distance) ** 2)
        restraint_info = {
            "target": target_domain_distance.tolist(),
            "restraint": restraint_domain_distance.tolist(),
            "prediction": pred_domain_distance.tolist(),
            "mse": float(mse),
        }
        add_arg_to_json(output_json, restraint_info)

        delta = target_domain_distance - pred_domain_distance
        adelta = np.abs(delta)
        mask = adelta > 1.0
        restraint_domain_distance[mask] = restraint_domain_distance[mask] + delta[mask]

        if mse < mse_threshold:
            np.savez_compressed(
                os.path.join(out_dir, name, f"seed_{seeds[0]}", "predictions", "inputs.npz"),
                **input_dict,
            )
            successful_cif_list.append(output_cif)
            break
        else:
            failed_cif_list.append(output_cif)

        seed += 1

    return successful_cif_list, failed_cif_list

native_pdb = "/content/drive/MyDrive/colab_cache/afmfold/storage/4G1E.pdb"
json_file = "/content/drive/MyDrive/colab_cache/afmfold-data/4G1E-add-msa.json"
out_dir = "/content/drive/MyDrive/colab_cache/afmfold-data/candidates"
name = "4G1E"

os.makedirs(out_dir, exist_ok=True)

native_traj = md.load(native_pdb)
domain_pairs = get_domain_pairs(name)

R = len(native_traj)
D = len(domain_pairs)
native_domain_distances = np.zeros((R, D))
for i, (d1, d2) in enumerate(domain_pairs):
    d = compute_domain_distance(native_traj, d1, d2)
    native_domain_distances[:, i] = d.ravel()
native_domain_distances *= 10.0

max_deviation = 50.0
grid_size = 5.0
min_dd = np.min(native_domain_distances, axis=0) - max_deviation
max_dd = np.max(native_domain_distances, axis=0) + max_deviation
num_half_grids = [math.ceil((max_dd[i] - min_dd[i]) / 2 / grid_size) for i in range(D)]
centers = [0.5 * (max_dd[i] + min_dd[i]) for i in range(D)]
grids_per_axis = [centers[i] + grid_size * np.arange(-num_half_grids[i], num_half_grids[i] + 1) for i in range(D)]
grids_per_axis = [g[g > 0] for g in grids_per_axis]
mesh_grids = np.meshgrid(*grids_per_axis)
grids = np.concatenate([mg.reshape((-1, 1)) for mg in mesh_grids], axis=-1)

mean_native_domain_distances = np.mean(native_domain_distances, axis=0)
delta = np.sum((grids - mean_native_domain_distances[None, :]) ** 2, axis=-1)
grids = grids[np.argsort(delta), :]

success_dict = {}
r_neighbors = 2
min_trial = 20

for i in tqdm(range(len(grids)), total=len(grids)):
    deltas = np.sum(np.abs(grids - grids[i][None, :]), axis=-1)
    neighbor_ind = np.nonzero(deltas < grid_size * r_neighbors)[0]
    if i >= min_trial and not any(j in success_dict and success_dict[j] for j in neighbor_ind):
        continue

    try:
        successful_cif_list, failed_cif_list = conditional_generation_local(
            target_domain_distance=grids[i],
            domain_pairs=domain_pairs,
            out_dir=out_dir,
            name=name,
            json_file=json_file,
            input_dict={},
        )
    except AssertionError as exc:
        print(f"Grid {i} failed before outputs were written: {exc}")
        successful_cif_list, failed_cif_list = [], []

    success_dict[i] = len(successful_cif_list) > 0

    for cif in failed_cif_list:
        failed_dir = Path(cif).resolve().parent.parent
        if failed_dir.exists():
            import shutil
            shutil.rmtree(failed_dir)

print(f"{sum(success_dict.values())} successful candidates written to {out_dir}/{name}")
```
